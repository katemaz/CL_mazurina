#### Задача 13.

Увеличить все встретившиеся в тексте четные числа на 1, а нечетные на 2, в том числе и отрицательные. Перед числом может стоять знак. Считаем, что между знаком и числом пробела нет.

```python
text_13 = '''Сегодня в городе было 12 градусов, но к вечеру температура упала до -5 градусов. По пути домой я заметил 7 машин, из которых 2 были синими. Я подумал, что это очень странно, ведь в этом районе раньше было не больше 4 синих машин. Я также встретил 13 человек, которые спешили по своим делам, но лишь 6 из них были в зимней одежде.'''

count = r'-?\d*[02468]'
uncount = r'-?\d*[03579]'

def count_mult(match):
    count_numb = int(match.group())
    return str(count_numb * 1)

def uncount_mult(match):
    uncount_numb = int(match.group())
    return str(uncount_numb * 2)

text_13_result = re.sub(count, count_mult, text_13)
text_13_result = re.sub(uncount, uncount_mult, text_13_result)
print(text_13_result)
```

```python
numbers = r'-?\d*[02468]|-?\d*[03579]'

def mult(match):
    numb = int(match.group())
    if numb % 2 == 0:
        return str(numb * 1)
    else:
        return str(numb * 2)
    return match.group()

text_13_multiplicated = re.sub(numbers, mult, text_13)
print(text_13_multiplicated)
```

3. Сделайте то же самое без регулярных выражений в одну строчку, используя bilt-in функцию filter.

```python
#1
text_14 = '''   Вчера температура была 12 градусов, а на следующий день -8.   

    Также я увидел 45 человек в очереди, и 205 автомобилей проехали мимо!         
         Сегодня прогноз обещает 15 градусов, но возможны осадки!     '''
spaces_1 = r'^\s+|\s+$'

def spaces(text_14):
    text_14_filtered = re.sub(spaces_1, '', text_14, flags=re.MULTILINE)
    return text_14_filtered
text_14_result = spaces(text_14)

print(text_14_result)
```

```python
#2
spaces_2 = r'^\s+$\n?|[\s\n\r]+'

text_14_2_result = re.sub(spaces_2, '', text_14, flags=re.MULTILINE)
print(text_14_2_result)
```
