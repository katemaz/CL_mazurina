#### Задача 2. 

Смоделируйте сценарий заказа еды в макдаке, определив четыре класса:

**Lunch** - класс, который управляет всем событием \
**Customer** - человек, который покупает блюдо \
**Employee** - продавец \
**Food** - сама еда

Какие у них должны быть методы:

    class Lunch:
        def __init__(self) - создает экземпляры покупателя и продавца
        def order(self, foodName) - начинает процесс заказа продавцом
        def result(self) - запрашивает у покупателя, какой у него экземпляр еды
        
    class Customer:
        def __init__(self) инициирует еду у покупателя значением None (у покупателя сперва нет еды, он ее еще не купил)
        def placeOrder(self, foodName, employee) - размещает заказ с экземпляром продавца, который обслуживает
        def printFood(self)  выводит название блюда
        
    class Employee:
        def takeOrder(self, foodName) - возвращает экземпляр еды с запрошенным названием
        
    class Food:
    def __init__(self, name) - сохраняет название блюда
    
Что должно происходить?
1. Конструктор класса Lunch заводит экземпляр покупателя и продавца, после чего метод order запрашивает у покупателя заказ, вызывая его метод placeOrder. В свою очередь placeOrder должен запрашивать у продавца новый объект Food, вызывая метод takeOrder. 
2. Объект Food должен хранить строку с названием блюда, переданную из Lunch.order() методу Customer.placeOrder(), методу Employee.takeOrder() и, наконец, конструктору Food. Наконец класс Lunch запрашивает о результате, который должен вывести, какую еду получил наш покупатель.

```python
class Lunch:
    def __init__(self):
        self.customer = Customer()
        self.employee = Employee()
   
    def order(self, foodName):
        self.customer.placeOrder(foodName, self.employee)
        
    def result(self):
        self.customer.printFood() 

class Customer:
    def __init__(self):
        self.food = None
    
    def placeOrder(self, foodName, employee):
        self.food = employee.takeOrder(foodName)
    
    def printFood(self):
        if self.food:
            print("Покупатель заказал: ", self.food.name)
        else:
            print("Покупатель ничего не заказал")

class Employee:
    def takeOrder(self, foodName):
        return Food(foodName)

class Food:
    def __init__(self, name):
        self.name = name

lunch = Lunch()
lunch.order(input('Введите заказ: '))
lunch.result()
```

#### Задача 3 (2 балла). 

Напишите игрушку "Тамагочи". У вас должно быть минимум три класса: два для питомцев, один для самой игры. При начале игры игрок должен выбирать, какое животное он хочет завести, и давать ему имя. Возраст и другие характеристики животного можно генерировать случайным образом. Животное можно кормить и играть с ним, еще оно спит. Сама игра должна проходить в цикле, на каждой итерации можно проверять, хочет ли животное спать, и отправлять его, все потребности можно реализовать через динамические атрибуты. Если животное долго не кормить, оно умирает. (Атрибуты можно обновлять на каждой итерации)

```python
import random

class Cat:
    def __init__(self, name):
        self.name = name 
        self.age = random.randint(1, 10)
        self.energy = random.randint(1, 30)
        self.hunger = random.randint(1, 30)
        self.clean = random.randint(1, 30)
        self.happiness = random.randint(1, 30)
        self.type = 'cat'

class Dog:
    def __init__(self, name):
        self.name = name 
        self.age = random.randint(1, 10)
        self.energy = random.randint(1, 30)
        self.hunger = random.randint(1, 30)
        self.clean = random.randint(1, 30)
        self.happiness = random.randint(1, 30)
        self.type = 'dog'

class Game:
    def check(self, pet):  #показатели питомца
        print(f'Показатели вашего питомца: ')
        print(f'Голод: {pet.hunger}')
        print(f'Чистота: {pet.clean}')
        print(f'Энергия: {pet.energy}')
        print(f'Настроение: {pet.happiness}')

        if pet.energy <= 10: 
            print(f'{pet.name} устал и спит.')
        elif pet.energy >= 25:
            print(f'{pet.name} полон сил и готов играть!')
        if pet.hunger >= 25:
            print(f'{pet.name} очень голоден!')
        elif pet.hunger <= 5:
            print(f'{pet.name} сыт!')
        elif pet.hunger == 30:
            print(f'Ваш питомец {pet.name} умер от голода.')

    def feed(self, pet):
        pet.hunger = max(0, pet.hunger - 7) #используем max для избежания отрицательных показаний
        pet.energy = min(30, pet.energy + 5) #используем min для избежания показаний > 30
        pet.happiness = min(30, pet.happiness + 3)
        print(f'Вы покормили питомца! {pet.name} сыт.')

    def play(self, pet):
        pet.happiness = min(30, pet.happiness + 5)
        pet.energy = max(0, pet.energy - 5)
        print(f'Вы поиграли с питомцем! {pet.name} счастлив!')

    def clean(self, pet):
        pet.clean = min(30, pet.clean + 10)
        pet.happiness = min(30, pet.happiness + 5)
        print(f'Вы убрались у питомца! {pet.name} доволен!')

    def hug(self, pet):
        pet.happiness = min(30, pet.happiness + 10)
        print(f'Вы обняли питомца! {pet.name} очень счастлив!')

pet_type = input('Выберите питомца: кот или собака: ')
name = input('Введите имя питомца: ')

if pet_type.lower() == 'кот':
    pet = Cat(name)
    print(f'Ваш питомец - кот по имени {name}')
elif pet_type.lower() == 'собака':
    pet = Dog(name)
    print(f'Ваш питомец - собака по имени {name}')
else:
    print('Некорректный выбор питомца.')
    exit()

game = Game()  #заводим экземпляр класса Game, чтобы использовать оттуда методы

while True:
    to_do = input('Выберите действие с питомцем: поиграть, покормить, помыть, обнять, показатели, завершить игру: ')
    if to_do.lower() == 'поиграть':
        game.play(pet)
    elif to_do.lower() == 'покормить':
        game.feed(pet)
    elif to_do.lower() == 'помыть':
        game.clean(pet)
    elif to_do.lower() == 'обнять':
        game.hug(pet)
    elif to_do.lower() == 'показатели':
        game.check(pet)
    elif to_do.lower() == 'завершить игру':
        print('Игра завершена.')
        break
```
