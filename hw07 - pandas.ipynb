#### Задание 11.

Выполните задание 10 с помощью циклов. Сравните время выполнения.

```python
start_time11 = time.time()

races_reading = {}
for group, score in df.groupby('race/ethnicity'):
    races_reading[group] = int(score['reading score'].mean())
print(f'Средний балл за экамен по чтению для каждой этнической группы: {races_reading}')


writing_min = {}
for group, score in df.groupby('parental level of education'):
    writing_min[group] = int(score['writing score'].mean())
print(f'\nМинимальный балл за экзамен по письму для каждого уровня образования: {writing_min}')

end_time11 = time.time()
```

```python
time10 = end_time10 - start_time10
time11 = end_time11 - start_time11

print(f'Время выполнения с помощью groupby: {time10}\nВремя выполнения с помощью циклов: {time11}')
```

#### Задание 13. 

Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
df['Math_PassStatus'] = df['math score'].apply(lambda x: 'F' if x < passmark else 'P')
df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'F' if x < passmark else 'P')
df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'F' if x < passmark else 'P')

math_passed = sum(df['Math_PassStatus'] == 'P')
math_failed = sum(df['Math_PassStatus'] == 'F')

reading_passed = sum(df['Reading_PassStatus'] == 'P')
reading_failed = sum(df['Reading_PassStatus'] == 'F')

writing_passed = sum(df['Writing_PassStatus'] == 'P')
writing_failed = sum(df['Writing_PassStatus'] == 'F')

print(f'Сдали математику: {math_passed}  Не сдали математику: {math_failed}\nСдали чтение: {reading_passed}  Не сделаи чтение: {reading_failed}\nСдали письмо: {writing_passed}  Не сдали письмо: {writing_failed} ')
```

Сколько студентов успешно сдали все экзамены?

#### Задание 15. 

Переведем баллы в оценки

*Система перевода баллов в оценки*

        [90;100] = A
        [80;90) = B
        [70;80) = C
        [60;70) = D
        [50;60) = E
        [0;50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    elif average_mark >= 80:
        return 'B'
    elif average_mark >= 70:
        return 'C'
    elif average_mark >= 60:
        return 'D'
    elif average_mark >= 50:
        return 'E'
    else:
        return 'F (Fail)'
df['average_mark'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

df['Grade'] = ((df['math score'] + df['reading score'] + df['writing score']) / 3).apply(GetGrade)

df.head(10)
```
